```python
import random
def fun():
    inside=0
    for i in range(n):
        x=random.uniform(-1,1)
        y=random.uniform(-1,1)
        if x**2+y**2<=1:
            inside+=1
    est=4*inside/n
    return est
print(f"estimate pi: {fun()}")
```
