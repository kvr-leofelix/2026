```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# basic graph
plt.plot([1,2,3],[10,20,25],label="1x",color="purple")
x=[1,2,3]
y=[10,20,30]
plt.plot(x,y,label="2x",linewidth=5,marker=".",linestyle="--",markersize=9,markeredgecolor='blue')
plt.plot([1,2,3],[12,20,33],label="3x")
plt.title("first graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel("x here")
plt.ylabel("y here")
plt.xticks([1,2,3,4,5])
plt.yticks([6,10,12,18,24,25,30,36])
plt.legend()
plt.show()
```

```python
x2=np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,color="red",linestyle="--")
plt.plot(x2[:5],x2[:5]**2,color="red",linewidth=5)
plt.plot(x2,x2**3,color="blue",linestyle="--")
plt.show()
```

```python
# resize the graph
plt.figure(figsize=(5,3),dpi=300)
a=np.arange(0,10,0.5)
plt.plot(a,a**4,label="this graph")
plt.legend()
plt.savefig("myfig.png",dpi=100)
plt.show()
```

```python
labels=['a','b','c']
values=[1,4,2]
plt.bar(labels,values)
br=plt.bar()


plt.figure(figsize=(4,4),dpi=50)
plt.show()
```

```python
labels=['a','b','c']
values=[1,4,2]
bar=plt.bar(labels,values)

bar[0].set_hatch('/')
bar[1].set_hatch('o')
bar[2].set_hatch('*')


plt.figure(figsize=(4,4),dpi=50)
plt.show()
```

```python
labels=['a','b','c']
values=[1,4,2]
bars=plt.bar(labels,values)
patter=['/','o','*']
for bar in bars:
    bar.set_hatch(patter.pop(0))

plt.figure(figsize=(4,4),dpi=50)
plt.show()
```
