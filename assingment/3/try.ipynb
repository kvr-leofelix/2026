```python
import numpy as np
import matplotlib.pyplot as plt
t=15
n=10
prior=0.3
belief=np.ones(n)*prior
pzio=0.85
pz1f=0.20
history=[]
np.random.seed(10)
for i in range(t):
    cell=np.random.randint(0,n)
    z=np.random.choice([0,1])
    po=belief[cell]
    pf=1-po
    if z==1:
        num = pzio*po
        dem= pzio*po + pz1f*pf
    else:
        num = (1 - pzio) * po
        dem = (1 - pzio) * po + (1 - pz1f) * pf
    belief[cell]=num/dem
    history.append(belief.copy())

plt.imshow(history,aspect='auto',cmap='hot')
plt.title("Occupancy Belief Evolution")
plt.colorbar(label="P(Obstacle)")
plt.show()
```
