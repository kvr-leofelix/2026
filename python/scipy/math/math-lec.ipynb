```python
x=4.2
y=3
base=10
print(math.ceil(x))#round up to nearest integer
print(math.floor(x))#round down to nearest integer
print(math.fabs(x))#retun the modulus value though maintaining the float
math.sqrt(x)
math.pow(x,y)#it is x**y
math.log(x)#normal log for base e
math.log(x,base)#log on base (base) -->exmath.log(1000,10)
# NOTE math.log(x,base)=math.log(x)/math.log(base)
math.log10(x)#similar to math.log(x,10)
```

```python
math.log10(1000)
math.log(1000,10)
# both give slight difference because of different way to proceed
```

```python
math.sin(x)
math.tan(x)
math.cos(x)
math.degrees(x)#convert radian to degree
math.radians(x)#convert degree to radian
math.hypot(x,y)#find the hypotenus of x,y
math.isclose(0.1+0.2,0.3)
math.factorial(y)
z=90
math.gcd(z,y)
n=10
k=2
math.comb(n,k)#nCk thisway you can choose among the k out of n
math.prod([2,4,5,6])
# or
ap=[1,2,3,34,21,4,5,6]
math.prod(ap)

```
