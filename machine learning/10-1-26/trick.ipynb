how to run the scatter for differt feacture to make plt with less labour

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load your data
df = pd.read_csv("insurance_data.csv")

# 2. Identify your features (all columns except the target)
# This creates a list of column names like ['age', 'income', 'score']
features = df.columns.drop('bought_insurance')

# 3. Loop through each feature and create a plot
for col in features:
    plt.figure(figsize=(6, 4)) # Creates a new window/space for each plot
    plt.scatter(df[col], df['bought_insurance'], marker='+', color='red')
    plt.title(f'Insurance Purchase vs {col}')
    plt.xlabel(col)
    plt.ylabel('Bought Insurance (0 or 1)')
    plt.show() # Forces the plot to display before moving to the next loop
```

this is the adv way to use saeborn and do the same above task

```python
import seaborn as sns

# This one line creates a grid of scatter plots for every combination!
# 'hue' colors the dots based on whether they bought insurance or not
sns.pairplot(df, hue='bought_insurance')
```

to make bar and hist out of many feacture with less labour

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load your data
df = pd.read_csv("HR-Employee-Attrition.csv")

# 2. Separate columns by type automatically
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# --- PLOT 1: Histograms for Numerical Data ---
print("Generating Histograms...")
for col in numerical_cols[:5]:  # Change [:5] to see more or all columns
    plt.figure(figsize=(10, 4))
    # 'hue' helps you see the comparison (e.g., who stayed vs who left)
    sns.histplot(data=df, x=col, hue='Attrition', kde=True, palette='magma')
    plt.title(f'Distribution of {col} by Attrition')
    plt.show()

# --- PLOT 2: Bar Charts (Countplots) for Categorical Data ---
print("Generating Bar Charts...")
for col in categorical_cols:
    if col == 'Attrition': continue # Skip the target column itself
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, hue='Attrition', palette='viridis')
    plt.title(f'Attrition Count by {col}')
    plt.xticks(rotation=45) # Rotate labels if they are long (like Job Roles)
    plt.show()
```
