```python
import numpy as np
import math
np.round()
math.isclose()
```

```python
print(np.round(2.5))
print(np.round(3.5))
print(np.round(3.1459,2))#upto two digit after decimal
ar=np.array([1.234,2.567,3.891])
print(np.round(ar,1))
```

```python
a=0.1+0.2
b=0.3
print(a==b)#False
print(math.isclose(a,b))#True
print(math.isclose(1.001,1.002,abs_tol=0.005))#output:True
# rel_tol:max allowed difference 
#abs_tol:min allowed difference
```
