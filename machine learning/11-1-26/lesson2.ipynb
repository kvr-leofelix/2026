```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
%time
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df
```

```python
plt.scatter(df.area,df.price,color="black",marker="+")
plt.xlabel("area")
plt.ylabel("price")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.predict([[1000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[4000]])
```

```python
a = pd.read_csv("area.csv")
a
```

```python
b= reg.predict(a)
# now store the output in the list a by this method
a['prices']=b
```

```python
a
```

```python
a.to_csv("prediction.csv",index=False)
# this makes a new csv file with the data in the onject
```
