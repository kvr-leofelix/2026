```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'Data': [1, 4, 9, 16, 25]})

# Use numpy's square root function on a pandas column
df['Sqrt'] = np.sqrt(df['Data'])

# Use numpy's log function
df['Log'] = np.log(df['Data'])

print(df)
```

```python
# Create a 'Status' column: 'High' if Data > 10, else 'Low'
df['Status'] = np.where(df['Data'] > 10, 'High', 'Low')
```

```python
# Create data with missing values
df['Values'] = [10, np.nan, 20, np.nan, 30]

# Fill NaN with the mean of the column using numpy
mean_val = df['Values'].mean()
df['Values'] = df['Values'].replace(np.nan, mean_val)
```

```python
# From DataFrame to NumPy Array
array_version = df[['Data', 'Sqrt']].to_numpy()

# From NumPy Array back to DataFrame
new_df = pd.DataFrame(array_version, columns=['A', 'B'])
```

```python
import numpy as np

data = np.array([10, 20, 20, 30, 40, 40, 50])

# Get unique values and their frequencies
vals, counts = np.unique(data, return_counts=True)

# Find the index of the maximum count
index = np.argmax(counts)

print(f"The mode is: {vals[index]}")
```
