```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
%time
```

```python
a = pd.read_csv("canada_per_capita_income.csv")
a.head()
```

```python
b = linear_model.LinearRegression()
b.fit(a[['year']],a.pci)
```

```python
plt.plot(a.year,a.pci,marker=".",color="red")
plt.plot(a.year,b.predict(a[['year']]),color="blue")               
```
