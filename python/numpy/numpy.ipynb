```python
a=np.array([1,2,3])
b=np.array([[1,2,3],[4,5,6]])
print(a.ndim,b.ndim)#tell dimensiom
print(a.shape,b.shape)#tell the (m,n) 
print(a.dtype,b.dtype)#tell the data type
print(a.itemsize,b.itemsize)#the size cover by the data entry in the array
print(a.size,b.size)#tells the number of element in the array
#note
# a.nbytes=a.size*a.itemsize #n.bytes tell the space taken by the data entry point
print(a.nbytes,b.nbytes)
```
